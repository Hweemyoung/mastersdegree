```python
print("2019\tS = %f\tp = %f"%(_s, _p))
_s2, _p2 = stats.ttest_ind(df2["Cited by"].dropna(), df2[df2.DOI.isin(_idx_papers2)]["Cited by"].dropna())
print("2014\tS = %f\tp = %f"%(_s2, _p2))
```

```python
db_handler.sql_handler.select("scopus_videos_2014_comp", "idx_paper").where("content", ["paper_explanation", "paper_assessment", "paper_application"], "in")
_records2_1 = db_handler.execute().fetchall()
temp = list(map(lambda _row:_row[0].split(", "), _records2_1))
_idx_papers2_1 = list()
for _row in temp:
    _idx_papers2_1 = _idx_papers2_1 + _row
_idx_papers2_1 = set(_idx_papers2_1)
len(_idx_papers2_1)
```

```python
plt.ylim([0, 500])
plt.boxplot([df2["Cited by"].dropna(), df2[df2.DOI.isin(_idx_papers2)]["Cited by"].dropna(), df2[df2.DOI.isin(_idx_papers2_1)]["Cited by"].dropna()],
            labels=[
                "2014 All\n(N=%s)"%len(df2["Cited by"].dropna()),
                "2014 w/ videos\n(N=%s)"%len(df2[df2.DOI.isin(_idx_papers2)]["Cited by"].dropna()),
                "2014 w/ dedicated videos\n(N=%s)"%len(df2[df2.DOI.isin(_idx_papers2_1)]["Cited by"].dropna())])
```

```python
print("2014 w/ videos\t\t\tS = %f\tp = %f"%(_s2, _p2))
_s2_1, _p2_1 = stats.ttest_ind(df2[df2.DOI.isin(_idx_papers2)]["Cited by"].dropna(), df2[df2.DOI.isin(_idx_papers2_1)]["Cited by"].dropna())
print("2014 w/ dedicated videos\tS = %f\tp = %f"%(_s2_1, _p2_1))
```

```python
plt.ylim([0, 100])
plt.boxplot([
    df1[~df1.DOI.isin(_idx_papers)]["Cited by"].dropna(),
    df1[df1.DOI.isin(_idx_papers)]["Cited by"].dropna(),
    df2[~df2.DOI.isin(_idx_papers2)]["Cited by"].dropna(),
    df2[df2.DOI.isin(_idx_papers2)]["Cited by"].dropna()
],
    labels=[
        "2019 w/o videos\n(N=%d)"%len(df1[~df1.DOI.isin(_idx_papers)]["Cited by"].dropna()),
        "2019 w/ videos\n(N=%d)"%len(df1[df1.DOI.isin(_idx_papers)]["Cited by"].dropna()),
        "2014 w/o videos\n(N=%d)"%len(df2[~df2.DOI.isin(_idx_papers2)]["Cited by"].dropna()),
        "2014 w/ videos\n(N=%d)"%len(df2[df2.DOI.isin(_idx_papers2)]["Cited by"].dropna())
    ])
```

```python
_s3, _p3 = stats.ttest_ind(
    df1[~df1.DOI.isin(_idx_papers)]["Cited by"].dropna(),
    df1[df1.DOI.isin(_idx_papers)]["Cited by"].dropna()
)
print("2019\tS = %f\tp = %f"%(_s3, _p3))
_s4, _p4 = stats.ttest_ind(
    df2[~df2.DOI.isin(_idx_papers2)]["Cited by"].dropna(),
    df2[df2.DOI.isin(_idx_papers2)]["Cited by"].dropna()
)
print("2014\tS = %f\tp = %f"%(_s4, _p4))
```
