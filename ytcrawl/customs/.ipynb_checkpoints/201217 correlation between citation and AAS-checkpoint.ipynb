```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df3 = pd.read_csv("/home/hweem/git/mastersdegree/ytcrawl/customs/scopus/scopus_2014_comp.csv")
```

```python
df3 = df3[df3["Cited by"].notna()]
```

```python
df3 = df3[df3["AAS"] != "None"]
```

```python
df3 = df3[df3["AAS"].notna()]
```

```python
df3 = df3[["AAS", "Cited by"]].astype(int)
plt.scatter(df3["AAS"], df3["Cited by"])
# plt.xlim(0, 10000)
# plt.ylim(0, 10000)
plt.xscale("log")
plt.yscale("log")
plt.xlabel("AAS")
plt.ylabel("Citation")
```

```python
np.corrcoef(df3["AAS"], df3["Cited by"])
```

```python
np.corrcoef(np.log10(df3["AAS"]), np.log10(df3["Cited by"]))
```

```python
def calc_plot(df):
    import numpy as np
    import matplotlib.pyplot as plt
    %matplotlib inline
    df = df[df["Cited by"].notna()]
    df = df[df["AAS"] != "None"]
    df = df[df["AAS"].notna()]
    df = df[["AAS", "Cited by"]].astype(int)
    plt.scatter(np.log10(df["AAS"]), np.log10(df["Cited by"]))
    plt.xlim(0, 4)
    plt.ylim(0, 4)
#     plt.xscale("log")
#     plt.yscale("log")
    plt.xlabel("log10(AAS)")
    plt.ylabel("log10(Citation)")
    print(np.corrcoef(np.log10(df["AAS"]), np.log10(df["Cited by"]))[0, 1])
```

```python
calc_plot(df3)
```

```python
df4 = pd.read_csv("/home/hweem/git/mastersdegree/ytcrawl/customs/scopus/scopus_2014_life.csv")
calc_plot(df4)
```

```python
df1 = pd.read_csv("/home/hweem/git/mastersdegree/ytcrawl/customs/scopus/scopus_2019_comp.csv")
calc_plot(df1)
```

```python
df2 = pd.read_csv("/home/hweem/git/mastersdegree/ytcrawl/customs/scopus/scopus_2019_life.csv")
calc_plot(df2)
```
